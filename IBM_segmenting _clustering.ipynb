### PART1: PREPARING THE DATAFRAME

```python
import pandas as pd
import wikipedia as wp

```

```python
html = wp.page("List of postal codes of Canada: M").html().encode("UTF-8")
df = pd.read_html(html)[0]
print (df.head())
```

```python
df.index -=1
```

```python
df.head()
```

```python
df.drop(-1 ,inplace =True)
df.rename(columns={0:'PostalCode', 1:'Borough' , 2:"Neighbourhood"} ,inplace =True)
```

```python
df.head()
```

```python
for i in range(0,len(df)):
 if df['Borough'][i] == 'Not assigned':
  df.drop(i,inplace=True)  

df
  
```

```python
df =df.reset_index(drop=True)
```

```python
df.head()
```

```python
group = df.groupby([ 'PostalCode','Borough'])["Neighbourhood"].apply(list)
```

```python
final=[]
por =[]
for i in range(0,len(group)):
 v = list(group.keys()[i])
 v.append(", ".join(group[i]))
 final.append(v)

```

```python
final_dataframe = pd.DataFrame(final)
```
